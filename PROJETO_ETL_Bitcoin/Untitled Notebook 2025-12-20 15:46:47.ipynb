```python
%sql
SELECT
  ValorBrl  AS last_price,
  timestamp AS last_timestamp
FROM pipeline_api_bitcoin.bitcoin_data.dados_bitcoin
  ORDER BY timestamp DESC
  LIMIT 1
```

```python
%sql
SELECT
  ValorBrl  AS max_price,
  timestamp AS max_timestamp
FROM pipeline_api_bitcoin.bitcoin_data.dados_bitcoin
  ORDER BY ValorBrl DESC, timestamp DESC
  LIMIT 1
```

```python
%sql
SELECT
  ValorBrl  AS min_price,
  timestamp AS min_timestamp
FROM pipeline_api_bitcoin.bitcoin_data.dados_bitcoin
  ORDER BY ValorBrl ASC, timestamp DESC
  LIMIT 1
```
